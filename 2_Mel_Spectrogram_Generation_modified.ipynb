```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
import librosa
from skimage.io import imsave
import librosa.display
```

```python
PATH="/content/drive/MyDrive/Dataset/vcc2016_training/SM2"

NO_FILES=len(os.listdir(PATH))
list_files={}
list_sr=[]
list_file=[]
count=0
n_fft = 2048
Dict_FFT={}
hop_length = 512
n_mels = 128
```

```python
print(NO_FILES)
```

```python
for i in os.listdir(PATH):
  
  name=os.path.join(PATH, i)
  Y, Sr=librosa.load(name)
  list_files["{}".format(name)]=Y
  list_sr.append(Sr)
  list_file.append(name)
  count+=1
 
```

```python
##Starting_Displaying_MelSpectrogram
count=0
for name in range(NO_FILES):
    
    
    X=list_files[list_file[name]]
    Y=list_sr[name]
    mel_spect = librosa.feature.melspectrogram(y=X, sr=Y, n_fft=n_fft, hop_length=hop_length)
    mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
    fig = plt.figure()
    
    librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time')
    count += 1
    plt.savefig('/content/drive/MyDrive/Output/SM2/mel_spectrogram/sm2_generated_mel_spectrogram'+str(count)+'.png')

    #plt.title('Mel Spectrogram');
    plt.colorbar(format='%+2.0f dB');
```
