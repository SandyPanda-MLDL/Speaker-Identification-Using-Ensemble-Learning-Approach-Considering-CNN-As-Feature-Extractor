```python
PATH="/home/sandipan/Videos/wav_files"
NO_FILES=len(os.listdir(PATH))
list_files={}
list_sr=[]
list_file=[]
count=0
n_fft = 2048
Dict_FFT={}
hop_length = 512
n_mels = 128


##Starting_Displaying_Amplitude_VS_Time

for name in os.listdir(path):
    
    Y, Sr=librosa.load(name)
    list_files["{}".format(name)]=Y
    list_sr.append(Sr)
    list_file.append(name)
    count+=1
    
for name in range(NO_FILES):
    
    X=list_files[list_file[name]]
    Y=list_sr[name]
    time = np.linspace(
        0, # start
        len(X) / Y,
        num = len(X)
    )
    
    
    plt.figure()
    
    plt.plot(time, X);
   
##Ending_Displaying_Amplitude_VS_Time




##Starting_Displaying_FFT

for name in range(NO_FILES):
    
    X=list_files[list_file[name]]
    FFT = np.abs(librosa.stft(X[:n_fft], n_fft=n_fft, hop_length=n_fft+1))
    Dict_FFT["{}".format(name)]=FFT
    plt.figure()
    plt.plot(FFT);
    
##Ending_Displaying_FFT




##Starting_Displaying_Spectrogram

for name in range(NO_FILES):
    
    X=list_files[list_file[name]]
    abs_amplitude = np.abs(librosa.stft(X, n_fft=n_fft,  hop_length=hop_length))
    DB = librosa.amplitude_to_db(abs_amplitude, ref=np.max)
    plt.figure()
    librosa.display.specshow(DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='log');
    plt.colorbar(format='%+2.0f dB');
    
##Ending_Displaying_Spectrogram


##Starting_Displaying_MelSpectrogram

for name in range(NO_FILES):
    
    
    X=list_files[list_file[name]]
    Y=list_sr[name]
    mel_spect = librosa.feature.melspectrogram(y=X, sr=Y, n_fft=n_fft, hop_length=hop_length)
    mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
    plt.figure()
    
    librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time');
    plt.title('Mel Spectrogram');
    plt.colorbar(format='%+2.0f dB');
    
    

```
